```python
import os
import pandas as pd
```

```python
df1 = pd.read_csv('chefmozaccepts.csv')
df2 = pd.read_csv('chefmozcuisine.csv')
df_combined1 = df1.merge(df2, on="placeID")

df3 = pd.read_csv('chefmozhours4.csv')
df_combined2 = df_combined1.merge(df3, on="placeID")

df4 = pd.read_csv('geoplaces2.csv')
df_combined3 = df_combined2.merge(df4, on="placeID")

df5 = pd.read_csv('rating_final.csv')
df_combined_final = df_combined3.merge(df5, on="placeID")



df_combined_final

```

```python
df_combined_final.to_csv('combined_dataset.csv', index=False)

```
